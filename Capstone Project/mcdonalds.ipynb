```python
%matplotlib inline

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from pandas import unique
import json
from pandas.io.json import json_normalize
```

```python
menu = pd.read_csv('menu.csv')
menu.sort_values('Total Fat',ascending=False).head()
```

```python
fig= plt.figure(figsize=(8,8))
ax1= fig.add_subplot(1,1,1)
for cat in menu['Category'].unique():
    
    x = menu['Calories'][menu['Category']==cat]
    y= menu['Total Fat'][menu['Category']==cat]
    ax1.plot(x,y,'o',color=np.random.uniform(size=3),label=cat)

plt.xlabel('Calories')
plt.ylabel('Total Fat')
handles, labels = ax1.get_legend_handles_labels()
plt.legend(handles=handles,loc='upper left')
plt.grid()
```
