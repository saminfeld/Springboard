```python
years = ['2010','2011','2012','2013','2014','2015','2016','2017']
weeks = np.arange(1,18)
```

```python
def scrape_teamstats(link1,allcolumns,floatcols,pergame,columnswanted):
    t0 = time.time()
    df_name = pd.DataFrame(columns = allcolumns)
    for y in years:
        link =  link1 + str(y) + "&seasonType=REG&tabSeq=2&qualified=false&Submit=Go"
        html = urlopen(link)
        html = BeautifulSoup(html.read(),'lxml')
        table = html.find('tbody').findAll('tr')

        for x in table:
            td = x.findAll('td')
            listy = []
            listy.append(y)
            count=len(df_name)
            for a in td:
                listy.append(a.text)
            listy= [re.sub('\n','',a) for a in listy]
            listy= [re.sub('\t','',a) for a in listy]
            df_name.loc[count,:]= listy
    
    #make columns floats
    floatcols = floatcols + ['G']
    for a in floatcols:
        df_name[a]=df_name[a].astype(float)
    
    #make some stats per game
    for a in pergame:
        df_name[a] = round(df_name[a] / df_name['G'],2)
    
    #get only columns wanted
    df_name = df_name[columnswanted]
        
    print(time.time() - t0)
    return df_name
```

```python
def scrape_schedule(link1):
    t0 = time.time()
    df = pd.DataFrame(columns = ['Year','Week','AwayTeam','AwayScore','HomeTeam','HomeScore'])
    for y in years:
        print(y)
        for w in weeks:
            count=0
            link =  link1 + str(y) + "/REG" + str(w)
            html = urlopen(link)
            html = BeautifulSoup(html.read(),'lxml')
            teamnames = html.findAll('span',{'class':'team-name'})
            teamscores = html.findAll('span',{'class':'team-score'})
            for a in np.arange(0,len(teamnames)/2):
                listy = []
                listy.append(y)
                listy.append(w)
                listy.append(teamnames[count].text)
                if len(teamscores)!=0:
                    listy.append(teamscores[count].text)
                else:
                    listy.append(0)
                listy.append(teamnames[count + 1].text)
                if len(teamscores)!=0:
                    listy.append(teamscores[count + 1].text)
                else:
                    listy.append(0)
                count +=2
                df.loc[len(df),:] = listy
    
    #make integers
    for c in ['Week','AwayScore','HomeScore']:
        df[c] = df[c].astype(int)
    
    print(time.time() - t0)
    return df     
```

```python
link1 = "http://www.nfl.com/schedules/"
copysched = scrape_schedule(link1)
schedules = copysched.copy()
schedules= schedules[['Year','Week','HomeTeam','HomeScore','AwayTeam','AwayScore']]
```

```python
#get cat codes
sched_cat = schedules.copy()
sched_cat['HomeTeam'] = sched_cat['HomeTeam'].astype('category')
sched_cat['AwayTeam'] = sched_cat['AwayTeam'].astype('category')
cat_columns = sched_cat.select_dtypes(['category']).columns
team_dict = dict(enumerate(sched_cat['HomeTeam'].cat.categories))
sched_cat[cat_columns] = sched_cat[cat_columns].apply(lambda x: x.cat.codes)

#conditions
train_cond = sched_cat['Year'] < 2017
test_cond = sched_cat['Year'] == 2017
#get train and test data
traindata = sched_cat.loc[train_cond]
testdata = sched_cat.loc[test_cond]

#Split X and y
X_train = traindata.drop(['HomeScore','AwayScore','Spread'],axis=1)
y_train = traindata[['AwayScore','HomeScore','Spread']]
X_test = testdata.drop(['AwayScore','HomeScore','Spread'],axis=1)
y_test = testdata[['AwayScore','HomeScore','Spread']]
```

```python
clf = DecisionTreeRegressor(min_samples_leaf=5)
clf.fit(X_train,y_train)
```
