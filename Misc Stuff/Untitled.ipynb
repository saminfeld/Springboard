```python
%matplotlib inline
import pandas as pd
import sklearn
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Setup Seaborn
sns.set_style("whitegrid")
sns.set_context("poster")
```

```python
war3 = pd.read_excel("./Warcraft3.xlsx", sheetname=0)
war3 = war3.reset_index()
del war3['index']

war3 = war3.replace('None',np.nan)
war3.fillna(0)
```

```python
def rid_slash(original,new1,new2):   
    others = pd.DataFrame(war3[original].str.split('/', 1, expand=True))
    others.columns=[new1,new2]
    war3[new1],war3[new2]= others[new1],others[new2]
    war3[new1]=war3[new1].fillna(war3[original])
    war3[new2]=war3[new2].fillna(war3[original])
rid_slash('Hit Points','HP1','HP2')
rid_slash('Sight','SightDay','SightNight')
rid_slash('Ground Attack','Grd Att1','Grd Att2')
war3
```
