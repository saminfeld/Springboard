## Get the Data

```python
t0 = time.time()
#edit weeks here
weeks_played = 3

#dataframe
df=[]
df= pd.DataFrame(columns=['League Member','Week','Position','Name','Opponent','Team Outcome','Points'])

#lists for searching

#secondlist


#links list and ranges
links = []
box = np.arange(1,13)
week = np.arange(1,weeks_played+1)


for w in week:
    for t in box:   
        link = ("http://games.espn.com/ffl/boxscorequick?leagueId=622044&teamId=" + str(t) + "&scoringPeriodId=" + str(w) + "&seasonId=2017&view=scoringperiod&version=quick")
        html = urlopen(link)
        html = BeautifulSoup(html.read(),'lxml')
        line1 = [html.find('table',{'id':'playertable_0'}).find('tr').text,w]

        thesearch = [html.findAll('table',{'id':'playertable_0'}),
                html.findAll('table',{'id':'playertable_1'})]

        #first players starters
        for i in np.arange(0,2):
            for a in thesearch[i]:
                #just the players
                for b in a.findAll('tr',{'class':'pncPlayerRow'}):
                    #list with team name already in it
                    line = line1.copy()
                    count=len(df)
                    #each individual data piece
                    for c in b.findAll('td'):
                        line.append(c.text)
                        if c.text=='** BYE **':
                            line.append('BYE')
                    df.loc[count,:]= line
time.time() - t0
```

## Clean the Data

```python
#fix points
if df['Points'].dtype!=float:
    df.loc[df['Points']=='--','Points']=0
df['Points']=df['Points'].astype(float)

#fix week
df['Week']=df['Week'].astype(int)

#make Win, Loss, Bye (WLB) column
df['TeamWL']=df['Team Outcome'].str[2]
df.loc[df['TeamWL']=='E','TeamWL']='B'

#get name and (*)
df['PName']=df['Name'].str.split(',',expand=True)[0]
#remove *
df.loc[df['PName'].str[-1]=='*','PName']=df.loc[df['PName'].str[-1]=='*','PName'].str[:-1]

#New Opponent
df['NewOpp']= df['Opponent'].map(lambda x: x.lstrip('@'))

#NewTeam
df['Team'] = df['Name'].str.split(',',expand=True)[1].str.split(expand=True)[0]
#Add Team for D/ST
df.loc[df['Team'].isnull(),'Team']=df['Name'].str.split(expand=True)[0]

#NewPosition
df['NewPos'] = df['Name'].str.split(',',expand=True)[1].str.split(expand=True)[1]
df.loc[df['NewPos'].isnull(),'NewPos']='D/ST'

#HomeAway
df['Home'] = 1
df.loc[df['Opponent'].str.contains('@'),'Home'] = 0

#LeagueMember
df['League Member']= df['League Member'].map(lambda x: x.replace(' Box Score',''))
```
